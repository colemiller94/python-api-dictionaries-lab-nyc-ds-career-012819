### Q2: This is only the first 50 results.  Get the next 50 and put them together.

```python

def brewery_getter(num_responses,state):
    url = 'https://api.openbrewerydb.org/breweries'
    per_page = 50
    current_page = 1
    params = {
        'per_page' : per_page,
        'page' : current_page,
        'by_state': state
    }
    dta = []
    while len(dta)<num_responses:
        raw = requests.get(url,params)
        dta.extend(raw.json())
        current_page += 1
    return dta
```
